```python
def PrintList(List):
    
    for l in range(len(List)):
        print(List[l], "\n")
    
    return(None)
```

```python
# Modified from Jonathan's code

def MatchSectionRE(SectionNbr):
    # Match the RE for section SectionNbr with the list of example headers of SectionNbr
    print("----------\nSection:", SectionNbr, "\n----------\n")
    NbrTestCases = len(HeaderDict[SectionNbr])
    print("Number of test cases = ", NbrTestCases, "\n")
    RE = regex1 + str(SectionNbr) + regex2 + f"{sections[SectionNbr]}"
    
    OKlist = []
    NOKlist = []
    
    for i in range(NbrTestCases):
        try:
            m = re.search(RE, HeaderDict[SectionNbr][i], flags=re.MULTILINE | re.IGNORECASE)
            if m : # found a match
                s = f"{i} : {HeaderDict[SectionNbr][i]} = {m}\n"
                OKlist.append(s)
            else : # no match found
                s = f"{i} : {HeaderDict[SectionNbr][i]}\n"
                NOKlist.append(s)
        except Exception as e:
            s = f"{HeaderDict[SectionNbr][i]} : {e}\n"
            NOKlist.append(s)
    
    
    print("Number OK = ", len(OKlist), " Number Failed = ", len(NOKlist), "\n")
    print("Regular Expression = \n\n", RE, "\n")
    print("OK :\n")
    PrintList(OKlist)
    print("\nFailed : \n")
    PrintList(NOKlist)
    
    return(None)
```

```python
for Section in range(1,17):
    MatchSectionRE(Section)
    print("\n")
```
